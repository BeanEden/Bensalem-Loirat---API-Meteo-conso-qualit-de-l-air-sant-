```python
cwd = os.getcwd()
cwd_data = cwd + "/data/"
cwd_file = cwd_data + "export_infoclimat.csv"
```

```python
df = pd.read_csv(cwd_file, delimiter=";")
df = df.drop([0])
df.head(5)
```
